```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']
```

```python
# european electricity, bj-traffic, synthetic data, space ocupition
tick_label = ['', '', '']

x = np.arange(3)
# y1 means raw, y2 means postgres, y3 means postgres-index, y4 means neo4j, y5 means neo4j-index, y6 means tgraph
# every element: eu, bj, synthetic
y1 = [4.5, 11.5, 8.0]
y2 = [14.8, 24.4, 30.5]
y3 = [16.2, 27.8, 37.1]
y4 = [9.3, 23.9, 31.3]
y5 = [9.5, 24.2, 33.9]
y6 = [8.7, 17.1, 20.4]


total_width, n = 0.84, 6
width = total_width / n
plt.bar(x, y1, width, fill=False, hatch='++', label='')
plt.bar(x+width, y2, width, fill=False, hatch='xx', label='PostgresSQL')
plt.bar(x+2*width, y3, width, fill=False, hatch='oo', label='PostgresSQL-Index')
plt.bar(x+3*width, y4, width, fill=False, hatch='//', label='Neo4j')
plt.bar(x+4*width, y5, width, fill=False, hatch='OO', label='Neo4j-Index')
plt.bar(x+5*width, y6, width, fill=False, hatch='..', label='TGraph')

plt.grid(color='#95a5a6', linestyle='--', linewidth=1.6, axis='y', alpha=0.7)

plt.legend();
plt.ylabel('GB')
plt.xticks(x+2.5*width, tick_label)
plt.show()
```
