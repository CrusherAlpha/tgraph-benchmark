```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']
```

```python
# micro bench, topo, throughput
tick_label = ['', '', '', '', '']

x = np.arange(5)
# y1 means postgres, y2 means neo4j, y3 means tgraph
y1 = [12, 7, 16, 13, 85]
y2 = [15, 12, 31, 28, 122]
y3 = [15, 12, 31, 28, 122]

total_width, n = 0.8, 3
width = total_width / n
# plt.bar(x, y1, width, color='black', label='PostgreSQL')
plt.bar(x, y1, width, fill=False, hatch='xx', label='PostgreSQL')
# plt.bar(x+width, y2, width, color='grey', label='Neo4j')
plt.bar(x+width, y2, width, fill=False, hatch='//', label='Neo4j')
# plt.bar(x+2*width, y3, width, color='gainsboro', label='TGraph')
plt.bar(x+2*width, y3, width, fill=False, hatch='..', label='TGraph')

plt.grid(color='#95a5a6', linestyle='--', linewidth=1.6, axis='y', alpha=0.7)


plt.legend()
plt.xticks(x+width, tick_label)
plt.ylabel('kops/s')
plt.show()
```
